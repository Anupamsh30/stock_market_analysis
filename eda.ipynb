## Stock Price List

```python
path_to_stock = '/Users/anupam.sharma/Documents/stock_market_analysis/data/indian_stock_data'
```

```python
#get path of all csv files
files = os.listdir(path_to_stock)
csv_files = glob.glob(path_to_stock + "/*.csv")

stock_name = [name.split('.')[0] for name in files]

stock_data_list = []
#read the data
for path in csv_files:
    filename = os.path.basename(path).split('.')[0]
    stock_single = pd.read_csv(path)
    stock_single['stock_name'] = str(filename)
    stock_data_list.append(stock_single)


# concatenate all stocks into one
stock_data   = pd.concat(stock_data_list, ignore_index=True)
```
